```python
def add_bias(array):
    ''' 
    This function returns a NumPy array with and extra column of ones.
    inputs:
    NumPy array
    
    outputs:
    NumPy array with an extra column
    '''
    array = np.hstack([array, np.ones((array.shape[0], 1))])
    return array
```

```python
X = add_bias(X)
X.shape
```

```python
def sigmoid(array):
    ''' 
    This function returns a NumPy array where all of the original values habe been
    transformed by the sigmoid logistic function.
    inputs:
    NumPy array
    outputs:
    NumPy array with transformed values by the Sigmoid function.
    '''
    
    s = 1/(1+(np.exp(-array)))
    return s
```

```python
X = sigmoid(X)
X
```

```python
a = np.array([-10.0, -1.0, 0.0, 1.0, 10.0])
expected = np.array([0.0, 0.27, 0.5, 0.73, 1.0])
assert np.all(sigmoid(a).round(2) == expected)
```

```python
def feed_forward(X, weights_h, weights_o):

    """
    1. Multiply the input matrix X
       with the weights of the first layer.

    2. Apply the sigmoid function on the result.

    3. Append an extra column of ones to the result (i.e. the bias).

    4. Multiply the output of the previous step
       with the weights of the second (i.e. outer) layer.

    5. Apply the sigmoid function on the result.

    6. Return all intermediate results (i.e. anything that is outputted
       by an activation function).
    """
    print('X with bias')
    #1. H_in
    weighted_sum = np.dot(X, weights_h)
    print(f'Shape of H_in = {weighted_sum.shape}')
    print('H_in')
    print(weighted_sum)
    
    #2. H_out
    output_1 = sigmoid(weighted_sum)
    print('H_out')
    print(output_1)
    
    #3. add_bias
    input_out = add_bias(output_1)
    print('H_out with bias')
    print(input_out)
    
    #4. O_in
    print(f'Shape of H_in = {weighted_sum.shape}')
    weighted_sum = np.dot(input_out, weights_o)
    print('O_in')
    print(weighted_sum)

    #5. O_out -> actual prediction
    output_2 = sigmoid(weighted_sum)
    print('O_out')
    print(output_2)

    return output_1, output_2
```

```python
out1, out2 = feed_forward(X, weights_h, weights_o)
```

```python
assert out1.shape == (500, 2)
assert out2.shape == (500, 1)
```

### Test

```python
out1, out2 = feed_forward(X, weights_h, weights_o)
assert out1.shape == (500, 2)
assert out2.shape == (500, 1)

Xref = np.array([[1.0, 2.0, 1.0]])
whidden = np.array([[1.0, 2.0, 0.0],
                 [-1.0, -2.0, 0.0]
                    ]).T
wout = np.array([[1.0, -1.0, 0.5]]).T

out1, out2 = feed_forward(Xref, whidden, wout)
assert np.all(out1.round(2) == np.array([[0.99, 0.01]]))
assert np.all(out2.round(2) == np.array([[0.82]]))
```
